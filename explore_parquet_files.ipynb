```python
import pandas as pd

# Lese die Parquet-Datei
df = pd.read_parquet("lichess_moves_final.parquet")

# Zeige die ersten 10 Zeilen an
print(df.head(10))
```

```python
print(df.loc[35, 'BaseFEN'])
```

```python
max_row = df.loc[df['TimesPlayed'].idxmax()]
print(max_row)
```

```python
#most played moves in a certain position, elo 1400-1699
df_test = df[df['BaseFEN'] == 'r1bqkbnr/ppp2ppp/3p4/4n3/4P3/5N2/PPPP2PP/RNBQKB1R w KQkq -']
df_test = df_test[(df_test['PlayerEloBucket'] >= 1400) & (df_test['PlayerEloBucket'] <= 1600)]

df_grouped = df_test.groupby('MoveSAN', as_index=False)['TimesPlayed'].sum()
df_grouped = df_grouped.sort_values(by='TimesPlayed', ascending=False).reset_index(drop=True)

print(df_grouped.head(5))
```
